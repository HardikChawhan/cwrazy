```python
# Step 1: Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Create sample data
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)  # y is the sine of x
y2 = np.random.randn(100)  # random data for scatter plot

# Step 3: Simple Line Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Sine Wave", color="b", linestyle="--", marker="o")  # Line plot with dashed line and circle markers
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()

# Step 4: Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y2, label="Random Data", color="r", marker="x")  # Scatter plot with red 'x' markers
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()

# Step 5: Histogram
data = np.random.randn(1000)  # Generate random data for histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color="green", edgecolor="black", alpha=0.7)  # Histogram with 30 bins
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Step 6: Bar Plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 25, 30]  # Values for the bars
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color="purple", alpha=0.7)  # Bar plot with purple color
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
