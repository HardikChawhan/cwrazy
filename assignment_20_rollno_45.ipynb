```python
# Step 1: Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Step 2: Load a dataset (We use Seaborn's built-in 'tips' dataset)
tips = sns.load_dataset('tips')

# Step 3: Create Subplots
fig, axs = plt.subplots(2, 3, figsize=(18, 12))  # 2 rows, 3 columns for the plots

# Step 4: Scatter Plot
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=axs[0, 0], color='blue')
axs[0, 0].set_title('Scatter Plot: Total Bill vs Tip')

# Step 5: Histogram
sns.histplot(tips['total_bill'], kde=True, ax=axs[0, 1], color='green', bins=20)
axs[0, 1].set_title('Histogram: Total Bill')

# Step 6: Bar Plot
sns.barplot(data=tips, x='day', y='total_bill', ax=axs[0, 2], color='purple')
axs[0, 2].set_title('Bar Plot: Total Bill by Day')

# Step 7: Box Plot
sns.boxplot(data=tips, x='day', y='total_bill', ax=axs[1, 0], palette='coolwarm')
axs[1, 0].set_title('Box Plot: Total Bill by Day')

# Step 8: Violin Plot
sns.violinplot(data=tips, x='day', y='total_bill', ax=axs[1, 1], inner="quart", palette='Set2')
axs[1, 1].set_title('Violin Plot: Total Bill by Day')

# Step 9: Pair Plot
sns.pairplot(tips[['total_bill', 'tip', 'size', 'age']], hue='sex', palette='coolwarm')
plt.suptitle('Pair Plot: Correlations between Variables', y=1.02)

# Step 10: Adjust layout and show plots
plt.tight_layout()
plt.show()
```
